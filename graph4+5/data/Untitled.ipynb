```python
import matplotlib.pyplot as plt
import numpy as np

def get_rewards_and_q(file):
	lines = open(file).readlines()
	rewards = []
	q_values = []
	for line in lines:
		reward, q_value = line.strip().split()
		reward = float(reward)
		q_value = float(q_value)
		# if q_value < 0 or reward > 90:
		# 	continue
		rewards.append(reward)
		q_values.append(q_value)
	return rewards, q_values

end_path = "mean_q_test_score.csv"
```

```python
def plot(dqn_rewards, dqn_q_values, ebu_rewards, ebu_q_values, dqn_label, ebu_label):
    # plt.xticks(np.arange(0, 200, 50))
# plt.yticks(np.arange(0, 10, 2))
# plt.axis([0, 10, 0, 200])
    plt.figure(figsize=(10,7))
    fig, ax = plt.subplots()
    plt.scatter(dqn_rewards, dqn_q_values, marker='o', label=dqn_label)
    plt.scatter(ebu_rewards, ebu_q_values, marker='x', label=ebu_label)
    plt.xticks(range(0, 300, 50))
    plt.yticks(range(0, 20, 2))
    plt.ylim(0, 10)
    plt.xlim(0, 200)
    ax.legend()
    ax.grid(True)
    plt.xlabel("Test Episode Score")
    plt.ylabel("Mean Q Values Across All Test Episodes")
    plt.show()
```

```python
dqn_rewards, dqn_q_values = get_rewards_and_q("mean_q_test_score_dqn.csv")
dqn_label = "DQN with Nature DQN Hyperparameters"
#############################
# EBU Beta=1 and DQN Parameters 
#############################
ebu_rewards, ebu_q_values = get_rewards_and_q("1ebu_"+end_path)
ebu_label = "Beta = 1 and Nature DQN Hyperparameters"
plot(dqn_rewards, dqn_q_values, ebu_rewards, ebu_q_values, dqn_label, ebu_label)
#############################
# EBU Beta=1 and EBU Parameters 
#############################
ebu_rewards, ebu_q_values = get_rewards_and_q("1"+end_path)
ebu_label = "Beta = 1 and EBU Hyperparameters"
plot(dqn_rewards, dqn_q_values, ebu_rewards, ebu_q_values, dqn_label, ebu_label)
#############################
# EBU Beta=0.5 and EBU Parameters 
#############################
ebu_rewards, ebu_q_values = get_rewards_and_q("5ebu"+end_path)
ebu_label = "Beta = 0.5 and EBU Hyperparameters"
plot(dqn_rewards, dqn_q_values, ebu_rewards, ebu_q_values, dqn_label, ebu_label)
#############################
# Nature DQN with EBU Parameters 
#############################
# dqn_rewards, dqn_q_values = get_rewards_and_q("last_"+end_path)
# ebu_label = "Beta = 1 and Nature DQN Hyperparameters"
# plot(dqn_rewards, dqn_q_values, ebu_rewards, ebu_q_values, dqn_label, ebu_label, title)

# ebu_rewards, ebu_q_values = get_rewards_and_q("mean_q_test_score_ebu_1.csv")
#ebu_n = len(ebu_rewards)
dqn_rewards = dqn_rewards #dqn_rewards[:ebu_n]
dqn_q_values = dqn_q_values #dqn_q_values[:ebu_n]

```
